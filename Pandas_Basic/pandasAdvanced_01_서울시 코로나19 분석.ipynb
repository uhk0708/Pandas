```python
raw = pd.read_csv('./data/서울시 코로나19 확진자 현황.csv')
df = raw.copy()
```

```python
# 데이터 확인 >> 전처리

# 컬럼별 데이터 확인
df.head()
```

```python
df.columns
```

```python
df['이동경로'].unique()
```

```python
# 불필요한 컬럼 삭제
df.drop(columns=['환자번호','국적', '환자정보', '조치사항','이동경로', '등록일', '수정일', '노출여부'], inplace=True)
df
```

```python
# 자료형 변환

# 자료형 확인
df.dtypes
```

```python
# 확진일 >> datetime
df['확진일'] = pd.to_datetime(df['확진일'])
df.dtypes
```

```python
# '지역' 공백 제거

df['지역'].nunique()
```

```python
df['지역'].unique()
```

```python
# 문자 양쪽 공백 제거 >> 업데이트
df['지역'] = df['지역'].str.strip()
print(df['지역'].nunique())
df['지역'].unique()
```

```python
# 지역 상태 >>  category (지역의 공백 제거)
df['지역'] = df['지역'].astype('category')
df.dtypes
```

```python
df.info()
```

```python
# 결측치 분석
df.isnull().sum()
```

```python
# 구 별 확진자 동향
# 확진일-구별 확진자 수 집계

# 피봇 테이블 만들기

df_gu = \
    pd.pivot_table(df,
               index='확진일',
               columns='지역',
               values='연번',
               aggfunc='count',
               margins=True)
```

```python
df_gu
```

```python
# 서울시 일별 추가 확진자 동향
df_gu['All']
```

```python
s_date = df_gu['All'][:-1]
```

```python
# 서울시 일별 추가확진자가 많았던 순으로 보기
s_date.sort_values(ascending=False)
```

```python
# 서울시 일별 추가 확진자 시각화

x = s_date.index # 일자
y = s_date.values # 확진자 수

plt.plot(x,y)
plt.title('서울시 일별 추가 확진자(2021.09.28 현재)')
plt.xlabel('확진일')
plt.ylabel('추가확진자 수')
plt.xticks(rotation=45)
plt.show()
```

```python
# 서울시 구 별 누적 확진자 비교
s_gu = df_gu.loc['All'][:-1]
```

```python
# 서울시 구 별 누적확진자가 많은 순으로 보기
s_gu = s_gu.sort_values(ascending=False)
s_gu
```

```python
# 서울시 구 별 누적확진자 시각화

x = s_gu.index # 서울시 구
y = s_gu.values # 누적 확진자 수

plt.figure(figsize=(10,6))
plt.title('서울시 구별 누적 확진자(2021.09.28 현재)', size=20)
plt.barh(x,y) #막대
plt.show()
```
