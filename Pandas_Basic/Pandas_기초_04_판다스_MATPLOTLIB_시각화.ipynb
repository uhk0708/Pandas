```python
# 그래프 그려보자
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y)
plt.title("My first Plot")

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title('My first Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')

fig.set_dpi(300)
# dpi (dot per inch) 해상도
# deprecated  다음 버전에서 안됨
fig.savefig('first_plot.png') # 그림 저장
# fig : figure 그림


plt.show()
```

```python
# 여러 개 그래프 그리기

np.linspace(0, np.pi*4, 100)

fig, axes = plt.subplots(2, 1)

axes[0].plot(x, np.sin(x))
axes[1].plot(x, np.cos(x))

plt.show()
```

Matplotlib 주요 그래프

```python
# 선 그래프 (line plot)

fig, ax = plt.subplots()

x = np.arange(15)
y = x ** 2

ax.plot(
    x, y, 
    linestyle = ":",
    marker="*",
    color="darkblue"
)

plt.show()
```

```python
# color

x = np.arange(10)

fig, ax = plt.subplots()

ax.plot(x,x,color="r") # r >> red
ax.plot(x,x+2,color="green")
ax.plot(x,x+4,color="0.8") # 0.8 >> 밝은 회색
# 그레이스케일 (흑백) 0~1 >> 0 : 검정색, 1 : 하얀색
ax.plot(x,x+6,color="#524FA1")


plt.show()
```

```python
# MARKER

x = np.arange(10)

fig, ax = plt.subplots()

ax.plot(x, x, marker=".")
ax.plot(x, x+2, marker='o')
ax.plot(x, x+4, marker='v')
ax.plot(x, x+6, marker='s')
ax.plot(x, x+8, marker='*')

plt.show()
```

```python
# 축 경계 조정하기

x = np.linspace(0, 10, 100)

fig, ax = plt.subplots()

ax.plot(x, np.sin(x))
ax.set_xlim(-2, 12)
ax.set_ylim(-2, 2)

plt.show()
```

```python
# 범례 (legend)


fig, ax = plt.subplots()

ax.plot(x, x, label='y=x')
ax.plot(x, x**2, label='y=x^2')

ax.set_title('legend of graph')
ax.set_xlabel('x')
ax.set_ylabel('y')

ax.legend(loc = 'best',
          #shadow = True,
          fancybox = True)
          #borderpad = 2 



plt.show()
```

```python
# 산점도 (scatter plot)

fig, ax = plt.subplots()

x = np.arange(10)
ax.plot(
    x, x**2, 'o',
    markersize=15,
    markerfacecolor='white',
    markeredgecolor='b')


plt.show()
```

```python
fig, ax = plt.subplots()

x = np.arange(10)
ax.scatter(
    x, x**2)


plt.show()
```

```python
# bubble 형태 scatter

fig, ax = plt.subplots()

x = np.random.randn(50)
y = np.random.randn(50)

colors = np.random.randint(0, 100, 50)
sizes = 500 * np.pi * np.random.rand(50) ** 2

ax.scatter(x,y,c = colors, s = sizes, alpha=0.3)
# alpha 투명도 (0~1 사이)

plt.show()
```

```python
# 바 그래프 (bar plot)

# bar

x = np.arange(10)

fig, ax = plt.subplots(figsize=(10,4))
ax.bar(x, x*2)

plt.show()
```

```python
# 누적 바 그래프

x = np.random.rand(3)
y = np.random.rand(3)
z = np.random.rand(3)

data = [x,y,z]


fig, ax = plt.subplots()

x_ax = np.arange(3)

for i in x_ax:
    ax.bar(x_ax, data[i],
           bottom = np.sum(data[:i], axis=0))
    
ax.set_xticks(x_ax)
ax.set_xticklabels(['A', 'B', 'C'])

plt.show()
```

```python
# Histogram(히스토그램)

fig, ax = plt.subplots()
data = np.random.randn(1000)



ax.hist(data, bins= 50) # bins : 계급구간
plt.show()
```
